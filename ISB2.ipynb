```python
#Initialising selenium
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(executable_path=r'C:\Users\CHITTI\Desktop\chromedriver.exe')
```

```python

#reading csv file
  

loc = (r"C:\Users\CHITTI\Desktop\link_by_areas.csv") 
  

df=pd.read_csv(loc)
print(df['link'][44])
```

```python
#to get the links from which the required o/p is to be generated
Link=[]
for i in range(0,45): 
    driver.get(df['link'][i])
    #timeDelay = random.randrange(0, 5)
    #time.sleep(timeDelay) 
    soup=BeautifulSoup(driver.page_source, 'lxml')#returns html of the page
    for i in soup.findAll(attrs={'class':"jobTuple bgWhite br4 mb-8"}):
        for j in i.findAll(attrs={'class':"title fw500 ellipsis"}):
            print(j.get('href'))
            Link.append(j.get('href')) #stores all the link of the job postings
```

```python
len(Link)
```

```python
#to get the required attributes.
salary=[]
experience=[]
Location=[]
description=[]
role=[]
industry_type=[]
qualification=[]
Functional_area=[]
Employment_type=[]
Role_category=[]
company=[]
skills=[]
posted=[] 


for lin in range(len(Link)):
    driver.get(Link[lin])
    #time.sleep(1)
    soup=BeautifulSoup(driver.page_source, 'lxml')
    #print(soup)
    if soup.find(attrs={'class':"salary"})==None: #to skip the error
        continue
    else:
        if soup.find(attrs={'class':"exp"})==None :
            experience.append('Not mentioned')
        else:
            experience.append(soup.find(attrs={'class':"exp"}).text)
        salary.append(soup.find(attrs={'class':"salary"}).text)
        if soup.find(attrs={'class':'loc'})==None :
            Location.append('Not mentioned')
        else:
            Location.append(soup.find(attrs={'class':'loc'}).find('a').text)

        
        description.append(soup.find(attrs={'class':"job-desc"}).text)
            

        details=[]

        for i in soup.find(attrs={'class':"other-details"}).findAll(attrs={'class':"details"}):
            details.append(i.text)
        details[0].replace('Role',' ')
        role.append(details[0])
        details[1].replace('Industry','').replace('Type', '')
        industry_type.append(details[1])
        details[2].replace('Functional','').replace('Area', '')
        Functional_area.append(details[2])
        details[3].replace('Employment','').replace('Type', '')
        Employment_type.append(details[3])
        details[4].replace('Role','').replace('Category', '')
        Role_category.append(details[4])

        qual=[]
        for i in soup.find(attrs={'class':"education"}).findAll(attrs={'class':'details'}):
            qual.append(i.text)
        qualification.append(qual)
        sk=[]
        if soup.find(attrs={'class':"key-skill"})==None:
            sk.append('No_value')
            skills.append(sk)
        else :
            
            for i in soup.find(attrs={'class':"key-skill"}).findAll('a'):
                sk.append(i.text)
            skills.append(sk)
        if soup.find(attrs={'class':"about-company"}).find(attrs={'class':"detail dang-inner-html"})==None:
            company.append('No_value')
            #else :
                #company.append(soup.find(attrs={'class':"about-company"}.text)
        else :                       
            company.append(soup.find(attrs={'class':"about-company"}).find(attrs={'class':"detail dang-inner-html"}).text)
```

```python

df=pd.DataFrame()
```

```python
#Transferring the data to a data-frame
df['company']=company
df['role']=role
df['salary']=salary
df['experience']=experience
df['Location']=Location
df['description']=description
df['skills']=skills
df['qualification']=qualification
df['industry_type']=industry_type

df['Functional_area']=Functional_area
df['Employment_type']=Employment_type
df['Role_category']=Role_category
```
